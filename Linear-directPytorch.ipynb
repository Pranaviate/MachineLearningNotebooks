```
y=np.array([[56.,70],
          [81,101],
          [119,133],
          [22,37],
          [103,119]])
```

```
y.dtype
```

```
targets=torch.from_numpy(y)
```

```
preds=x
```

```
print(x)
print(y)
```

```
print(preds)
print(targets)
```

```
#Linear Regression Model
```

```
# here 3 signifies number of components that effect the model for every statement
# 2 signifies number of result each row leads to
w=torch.randn(2,3, requires_grad=True)
```

```
# 2 signifies number of result each row leads to
b=torch.randn(2, requires_grad=True)
```

```
#Generate model
def model(a):
  return a @ w.t() +  b
```

```
#Generate Predictions
pred=model(preds)
print(pred)
```

```
print(preds)
```

```
#Loss function
diff=(pred-targets)
```

```
print(diff)
```

```
torch.sum(diff*diff)/diff.numel()
```

```
def mse(t1,t2):
  diff=t1-t2
  return torch.sum(diff*diff)/diff.numel()
```

```
loss=mse(pred,targets)
loss.backward()
```

```
print w
```

```
print(w)
```

```
print(w.grad)
```

```
#Optimization Algorithm
#As slope is positive here->>>
#We need to decrease the weight

with torch.no_grad():
  w-=w.grad*1e-5
```

```
w.grad.zero_()
b.grad.zero_()
```

```
#Creating an automated model
#for 100 Epochs
for i in range(100):
  pred=model(preds)
  loss=mse(pred,targets)
  loss.backward()
  with torch.no_grad():
    w-=w.grad*1e-5
    b-=b.grad*1e-5
    w.grad.zero_()
    b.grad.zero_()
  if i%10==0:
    print('epoch:',i)
    print('loss:',loss)
```
