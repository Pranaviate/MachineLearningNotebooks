```
tdata=pd.read_csv('Iris.csv')
```

```
feature_names=['Id','SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
x=pd.DataFrame(tdata, columns=feature_names)
#y=data['Species']
```

```
x.head()
```

```
y=pd.DataFrame(tdata, columns=['Species'])
```

```
from sklearn.tree import DecisionTreeClassifier

model=DecisionTreeClassifier()
model.fit(x,y)

from sklearn.tree import export_graphviz
export_graphviz(model,'tree.dot', feature_names=feature_names)
```
