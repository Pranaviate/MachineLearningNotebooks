# Demographic Filtering
This is done to balance the value of rating and number of votes to decide a better movie

### Weighted Rating(WR) = [v/(v+m)]*R+[m/(v+m)]*C

v=number of votes

m=minimum votes required for listing

R=average rating

C=mean vote

```python
C = df2['vote_average'].mean()
C
```

```python
m = df2['vote_count'].quantile(0.9)
m
```

```python
q_movs=df2.copy().loc[df2['vote_count'] >=m]
q_movs.shape
```

```python
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    # Calculation based on IMDB formula
    return (v/(v+m)*R) + (m/(m+v) * C)
```

```python
#adding new 'score' feature to qmovs
q_movs['score'] = q_movs.apply(weighted_rating, axis=1)
```

```python
q_movs = q_movs.sort_values('score', ascending=False)
q_movs.head(15)
```
